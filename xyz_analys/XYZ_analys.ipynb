# **XYZ-анализ**

```python
import pandas as pd
import numpy as np
```

```python
pd.set_option('display.max_columns', 40)
pd.set_option('display.float_format', '{:.2f}'.format)
```

```python
df = pd.read_csv('data1.csv', encoding='1251')
```

```python
df 
```

```python
df = df.groupby(['DR_Dat', 'DR_NDrugs']).agg({'DR_Kol': sum})
df
```

```python
df1 = df.reset_index().groupby('DR_NDrugs').agg({'DR_Dat': 'count'}).reset_index()
df1
```

## Фильтрация товаров, отвечающих требованиям

```python
xyz_names = list(df1[df1['DR_Dat'] > 1]['DR_NDrugs'])
xyz_names
```

```python
df = df.reset_index()
df = df[df['DR_NDrugs'].isin(xyz_names)]
df = df.drop(columns = ['DR_Dat'])
df = df.groupby('DR_NDrugs').apply(lambda x: x.std() / x.mean())
df.reset_index()
```

## Классификация товаров по признакам

```python
df['XYZ'] = np.where(df['DR_Kol'] < 0.1, 'X', np.where(df['DR_Kol'] < 0.25, 'Y', 'Z'))
```

```python
df = df.reset_index()
df
```
