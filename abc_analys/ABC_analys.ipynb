# Одномерный ABC анализ

```python
import pandas as pd
import numpy as np
```

```python
pd.set_option('display.max_columns', 40)
pd.set_option('display.float_format', '{:.2f}'.format)
```

```python
df = pd.read_csv('data/data.csv', encoding='1251')
```

```python
df.head(2)
```

```python
groupped_df = df.groupby('DR_NDrugs').agg({'DR_Kol': sum})
groupped_df
```

```python
groupped_df = (groupped_df['DR_Kol'] / sum(groupped_df['DR_Kol'])).sort_values(ascending=False)
groupped_df
```

## Расчет накопительной суммы

```python
groupped_df = groupped_df.cumsum().reset_index()
groupped_df
```

## Инициализация групп

```python
groupped_df['abc'] = np.where(groupped_df['DR_Kol'] < 0.8, 'A', np.where(groupped_df['DR_Kol'] < 0.95, 'B', 'C'))
```

```python
groupped_df[groupped_df['abc'] == 'B']
```
